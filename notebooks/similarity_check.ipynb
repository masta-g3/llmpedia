# Similarity Check
Identify documents with highly similar sections.

```python
import os, sys
import json
import pandas as pd
import numpy as np
from dotenv import load_dotenv
load_dotenv()
sys.path.append(os.environ.get("PROJECT_PATH"))

import utils.paper_utils as pu
data_path = os.path.join(os.environ.get("PROJECT_PATH"), "data")
```

```python
## Load data.
dict_names = os.listdir(os.path.join(data_path, "summaries"))
dict_names = [x for x in dict_names if x.endswith(".json")]
summaries = [json.load(open(os.path.join(data_path, "summaries", x))) for x in dict_names]
summaries = [pu.flatten_dict(x) for x in summaries]
summary = summaries[0]
```

```python
for summary in summaries:
    sim_matrix = pu.dict_similarity_matrix(summary)
    high_similarity_pairs = pu.get_high_similarity_pairs(sim_matrix, 0.99)
    if len(high_similarity_pairs) > 0:
        for tuple_names, sim in high_similarity_pairs:
            print(f"Similarity: {sim}")
            print(tuple_names[0])
            print(summary[tuple_names[0]])
            print("")
            print(tuple_names[1])
            print(summary[tuple_names[1]])
            print("=======================")
        print("....../..../..././"*2)
```
